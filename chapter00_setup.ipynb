## 0.1 我们要做什么？

想象你是一家新闻网站的编辑，每天有成千上万篇文章涌进来，你需要给每篇文章贴标签：🌍 World / ⚽ Sports / 💼 Business / 🔬 Science/Tech。

手动分类 128,600 篇？那要累死人。**我们的目标：让计算机自动学会这件事。** 这就是 **机器学习（Machine Learning）** 要做的事。

## 0.2 为什么用 Python？

Python 是全世界数据科学和机器学习领域最流行的编程语言。类比：Python 就像一把**瑞士军刀**，本身不大，但插上各种「工具」（库）之后几乎无所不能。

- 语法接近自然语言，容易上手
- 有海量现成工具，复杂算法别人已经写好
- 社区庞大，遇到问题 Google 一下总有答案

## 0.3 什么是 Google Colab？

**Google Colab** 是一个在浏览器里运行 Python 代码的免费平台。

- ✅ 不需要安装任何软件
- ✅ 免费提供 GPU（用于后面的深度学习）
- ✅ 代码和文字可以混在一起，边读边练
