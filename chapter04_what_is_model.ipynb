## 4.1 什么是模型（Model）？

**模型** = 一套从输入到输出的映射规则。

生活类比：
- 医生看症状（输入）→ 判断是什么病（输出）：医生的「经验」就是模型
- 银行看你的信用记录（输入）→ 决定是否贷款（输出）：审核规则就是模型
- 我们的任务：看文章内容（输入）→ 判断是哪类新闻（输出）

更准确地说：**模型是一个函数**，接收数据，输出预测。

```
输入（文章文本） → [模型] → 输出（World / Sports / Business / Science/Tech）
```

## 4.2 什么是训练（Training）？

**训练** = 让机器从数据里自动总结规律。

类比：学生刷题
1. 看一道题和标准答案（输入 + 正确标签）
2. 做出自己的答案
3. 对比标准答案，发现自己哪错了
4. 调整思路（更新参数）
5. 反复练习，做对率越来越高

机器也是这样：
1. 把有标签的文章喂给模型
2. 模型做出预测
3. 和正确标签对比，计算误差
4. 根据误差调整模型内部的参数
5. 反复迭代，预测越来越准

## 4.3 什么是预测（Prediction / Inference）？

**预测** = 用训练好的模型处理新数据。

类比：期末考试
- 训练阶段：用练习题学习
- 预测阶段：用学到的知识做考试题（考试题是没见过的新题）

**关键：** 预测时用的数据，必须是模型在训练时从未见过的。

## 4.4 什么是过拟合（Overfitting）？

**过拟合** = 模型把训练集「背」下来了，但遇到新数据就不行了。

类比：
- 有个学生只背答案，不理解题目。模拟考试（训练集）全对，期末考试（测试集）完全不会做。
- 这叫「背题」，不叫「学会」。

**为什么要测试集？**
- 测试集是模型从未见过的数据
- 在测试集上评估，才能知道模型是真的学会了，还是只是「背题」
- 测试集绝对不能参与训练！

## 4.5 为什么选这四个模型？

这个作业选了 4 个模型，从简单到复杂：

| 模型 | 类型 | 为什么选它 |
|------|------|----------|
| **朴素贝叶斯（Naive Bayes）** | 传统机器学习 | 最简单，速度最快，作为「基准线」 |
| **k 近邻（kNN）** | 传统机器学习 | 直觉最清晰，用「找相似」来分类 |
| **卷积神经网络（CNN）** | 深度学习 | 擅长检测局部文字模式（短语）|
| **双向 LSTM（BiLSTM）** | 深度学习 | 擅长理解长距离上下文和语序 |

**四个模型覆盖了两大类方法：**
- 传统机器学习：基于统计，速度快，可解释性强
- 深度学习：基于神经网络，更强大，需要更多数据和计算资源

## 4.6 sklearn 是什么？

**scikit-learn**（简称 sklearn）是 Python 最流行的传统机器学习库。

特点：
- 已经实现了几十种算法（朴素贝叶斯、kNN、SVM、随机森林等）
- **统一接口：** 所有模型用法相同：
  ```python
  model.fit(X_train, y_train)    # 训练
  model.predict(X_test)          # 预测
  model.score(X_test, y_test)    # 评估
  ```
- 不需要了解算法细节，直接用

> 官方文档：https://scikit-learn.org/stable/

## 4.7 TensorFlow / Keras 是什么？

**TensorFlow** 是 Google 开发的深度学习框架，用于构建和训练神经网络。

**Keras** 是 TensorFlow 的高层接口，像搭积木一样搭建神经网络：

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(128, activation='relu'),  # 第一层：128 个神经元
    Dense(4, activation='softmax')  # 输出层：4 个类别
])
model.compile(optimizer='adam', loss='categorical_crossentropy')
model.fit(X_train, y_train, epochs=10)
```

> 官方文档：https://www.tensorflow.org/tutorials

````python
# 用 sklearn 做一个简单演示：无需理解细节，感受一下流程
from sklearn.datasets import make_classification
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 生成一个假数据集（100 个样本，2 个特征，2 个类别）
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, random_state=42)

# 划分训练集和测试集
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3, random_state=42)

# 三步走：创建模型 → 训练 → 预测
model = GaussianNB()       # 1. 创建模型
model.fit(X_tr, y_tr)      # 2. 训练（fit = 拟合）
y_pred = model.predict(X_te)  # 3. 预测

print(f'测试集准确率：{accuracy_score(y_te, y_pred):.1%}')
print('这就是机器学习的完整流程！')
````

## 总结

| 概念 | 含义 |
|------|------|
| 模型（Model） | 从输入到输出的映射规则（函数）|
| 训练（Training） | 让模型从有标签的数据里自动学习规律 |
| 预测（Prediction） | 用训练好的模型处理新数据 |
| 过拟合（Overfitting） | 模型背答案，遇到新数据失效 |
| sklearn | 传统机器学习算法库，统一的 fit/predict 接口 |
| TensorFlow/Keras | 深度学习框架，像搭积木一样搭神经网络 |
